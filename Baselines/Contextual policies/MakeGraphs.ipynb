```python
def make_graphs_car():
    repeats = 5
    dicts = 10
    iters = 200
    dlistB = []
    for i in range(dicts):
        try:
            dlistB.append(load_obj("bc_values"+str(i),'Autonomous driving simulation/'))
        except:
            pass

    dlistC = []
    for i in range(dicts):
        try:
            dlistC.append(load_obj("coirl_values"+str(i),'Autonomous driving simulation/'))
        except:
            pass

    TBSTS = True
    cont_vec = [5,10,20,30,40,60,80,100]
    accurs_mean = []
    values_mean = []
    train_accurs_mean = []
    train_values_mean = []
    accurs_std = []
    values_std = []
    train_accurs_std = []
    train_values_std = []

    for d in dlistB:
        train_agent_values = []
        train_agent_accurs = []
        test_agent_values = []
        test_agent_accurs = []
        
        for trainset in range(repeats):
            test_agent_values.append([])
            test_agent_accurs.append([])
            train_agent_values.append([])
            train_agent_accurs.append([])
            cont_vecB = d["context_nums"]
            test_expert_value = d["test_value_expert"]
            train_expert_values = d["train_value_expert"]
            TBSTS = True
            for contexts in cont_vecB:
                test_agent_accurs[-1].append(d[trainset,"accuracy",contexts])
                test_agent_values[-1].append(d[trainset,"test_value",contexts])
                train_agent_accurs[-1].append(d[trainset,"train_accuracy",contexts])
                train_agent_values[-1].append(d[trainset,"train_value",contexts])

            accurs_mat = np.zeros([len(cont_vecB),repeats])
            values_mat = np.zeros([len(cont_vecB),repeats])
            train_accurs_mat = np.zeros([len(cont_vecB),repeats])
            train_values_mat = np.zeros([len(cont_vecB),repeats])
            
        for cont_i in range(len(cont_vecB)):
            for trainset in range(repeats):
                accurs_mat[cont_i,trainset] = test_agent_accurs[trainset][cont_i][0]
                values_mat[cont_i,trainset] = test_agent_values[trainset][cont_i][0]/test_expert_value
                train_accurs_mat[cont_i,trainset] = train_agent_accurs[trainset][cont_i]
                train_values_mat[cont_i,trainset] = train_agent_values[trainset][cont_i]/train_expert_values[trainset*len(cont_vecB) + cont_i]
                
        accurs_mean.append(accurs_mat.mean(axis=1))
        values_mean.append((100*values_mat).mean(axis=1))
        train_accurs_mean.append(train_accurs_mat.mean(axis=1))
        train_values_mean.append((100*train_values_mat).mean(axis=1))
        accurs_std.append(accurs_mat.std(axis=1))
        values_std.append((100*values_mat).std(axis=1))
        train_accurs_std.append(train_accurs_mat.std(axis=1))
        train_values_std.append((100*train_values_mat).std(axis=1))
    
    TCSTS = True   
    accurs_meanC = []
    values_meanC = []
    train_accurs_meanC = []
    train_values_meanC = []
    accurs_stdC = []
    values_stdC = []
    train_accurs_stdC = []
    train_values_stdC = []
    
    for d in dlistC:
        train_agent_valuesC = []
        train_agent_accursC = []
        test_agent_valuesC = []
        test_agent_accursC = []
        states_visitedC = []

        for trainset in range(repeats):
            test_agent_valuesC.append([])
            test_agent_accursC.append([])
            train_agent_valuesC.append([])
            train_agent_accursC.append([])
            states_visitedC.append([])
            test_expert_valueC = d["test_value_expert"]
            train_expert_valuesC = d["train_value_expert"]
            TCSTS = True
            
            for contexts in cont_vec:
                test_agent_accursC[-1].append(d[trainset,"accuracy",contexts])
                test_agent_valuesC[-1].append(d[trainset,"test_value",contexts])
                train_agent_accursC[-1].append(d[trainset,"train_accuracy",contexts])
                train_agent_valuesC[-1].append(d[trainset,"train_value",contexts])
                states_visitedC[-1].append(d[trainset,"states_visited",contexts])

            accurs_matC = np.zeros([len(cont_vec),repeats])
            values_matC = np.zeros([len(cont_vec),repeats])
            train_accurs_matC = np.zeros([len(cont_vec),repeats])
            train_values_matC = np.zeros([len(cont_vec),repeats])
            states_visited_matC = np.asarray(states_visitedC)
            
        for cont_i in range(len(cont_vec)):
            for trainset in range(repeats):
                accurs_matC[cont_i,trainset] = test_agent_accursC[trainset][cont_i][0]
                values_matC[cont_i,trainset] = test_agent_valuesC[trainset][cont_i][0]/test_expert_valueC
                train_accurs_matC[cont_i,trainset] = train_agent_accursC[trainset][cont_i]
                train_values_matC[cont_i,trainset] = train_agent_valuesC[trainset][cont_i]/ train_expert_values[trainset*len(cont_vecB) + 1]
                
        cont_vecC = states_visited_matC.mean(axis=0)[:-2]
        accurs_meanC.append(accurs_matC.mean(axis=1)[:-2])
        values_meanC.append((100*values_matC).mean(axis=1)[:-2])
        train_accurs_meanC.append(train_accurs_matC.mean(axis=1)[:-2])
        train_values_meanC.append((100*train_values_matC).mean(axis=1)[:-2])
        accurs_stdC.append(accurs_matC.std(axis=1)[:-2])
        values_stdC.append((100*values_matC).std(axis=1)[:-2])
        train_accurs_stdC.append(train_accurs_matC.std(axis=1)[:-2])
        train_values_stdC.append((100*train_values_matC).std(axis=1)[:-2])
    
    fig1 = plt.figure(figsize=(6,4.5))
    p1 = fig1.add_subplot(111)
    cont_vecB = np.asarray(cont_vecB)
    cont_vecB = (100/1531) * cont_vecB
    cont_vecC = (100/1531) * cont_vecC
    p1.plot(cont_vecB,values_mean[0], color='r')
    p1.plot(cont_vecC,values_meanC[0], color='b')
    p1.errorbar(cont_vecC,train_values_meanC[0],yerr=train_values_stdC[0], fmt='none', color='b',capsize=2)
    p1.scatter(cont_vecB,train_values_mean, marker="o", facecolors='none', edgecolors='r', label="BC train / test")
    p1.errorbar(cont_vecB,train_values_mean[0],yerr=train_values_std[0], fmt='none', color='r',capsize=5)
    p1.scatter(cont_vecC,train_values_meanC, marker="o", facecolors='none', edgecolors='b', label="COIRL (PSGD) train / test")
    p1.fill_between(cont_vecB,values_mean[0] - values_std[0], values_mean[0] + values_std[0], color='r', alpha=0.1)
    p1.fill_between(cont_vecC,values_meanC[0] - values_stdC[0], values_meanC[0] + values_stdC[0], color='b', alpha=0.1)
    p1.set_title("Value X Observability")
    p1.set_xlabel("% Policy seen")
    p1.set_ylabel("% of Expert value")
    p1.legend(loc=4)
    plt.xlim([0,30])
    
    fig2 = plt.figure(figsize=(6,4.5))
    p2 = fig2.add_subplot(111)
    p2.plot(cont_vecB,accurs_mean[0], color='r')
    p2.plot(cont_vecC,accurs_meanC[0], color='b')
    p2.errorbar(cont_vecC,train_accurs_meanC[0],yerr=train_accurs_stdC[0], fmt='none', color='b',capsize=2)
    p2.scatter(cont_vecB,train_accurs_mean, marker="o", facecolors='none', edgecolors='r', label="BC train / test")
    p2.errorbar(cont_vecB,train_accurs_mean[0],yerr=train_accurs_std[0], fmt='none', color='r',capsize=5)
    p2.scatter(cont_vecC,train_accurs_meanC, marker="o", facecolors='none', edgecolors='b', label="COIRL (PSGD) train / test")
    p2.fill_between(cont_vecB,accurs_mean[0] - accurs_std[0], accurs_mean[0] + accurs_std[0], color='r', alpha=0.1)
    p2.fill_between(cont_vecC,accurs_meanC[0] - accurs_stdC[0], accurs_meanC[0] + accurs_stdC[0], color='b', alpha=0.1)
    p2.set_title("Accuracy X Observability")
    p2.set_xlabel("% Policy seen")
    p2.set_ylabel("% Hit")
    p2.legend(loc=4)
    plt.xlim([0,30])
    plt.show()
    
```

```python
def make_graphs_grid():
    repeats = 5
    dicts = 10
    iters = 200
    dlistB = []
    for i in range(dicts):
        try:
            dlistB.append(load_obj("bc_values"+str(i),'Grid world/'))
        except:
            pass
    dlistBlin = []
    for i in range(dicts):
        try:
            dlistBlin.append(load_obj("bc_lin_values"+str(i),'Grid world/'))
        except:
            pass

    dlistC = []
    for i in range(dicts):
        try:
            dlistC.append(load_obj("coirl_values"+str(i),'Grid world/'))
        except:
            pass
    dim_prob = ["3X3","4X4","5X5","6X6","7X7","8X8","9X9","10X10"]

    TBSTS = True
    cont_vec = []
    cont_veclin = []
    cont_vecC = []
    accurs_mean = []
    values_mean = []
    train_accurs_mean = []
    train_values_mean = []
    accurs_std = []
    values_std = []
    train_accurs_std = []
    train_values_std = []
                   
    for d in dlistB:
        train_agent_values = []
        train_agent_accurs = []
        test_agent_values = []
        test_agent_accurs = []
        cont_vec = d["context_nums"]
        for trainset in range(repeats):
            test_agent_values.append([])
            test_agent_accurs.append([])
            train_agent_values.append([])
            train_agent_accurs.append([])
            test_expert_value = d["test_value_expert"]
            train_expert_values = d["train_value_expert"]
            TBSTS = True
            for contexts in cont_vec:
                test_agent_accurs[-1].append(d[trainset,"accuracy",contexts])
                test_agent_values[-1].append(d[trainset,"test_value",contexts])
                train_agent_accurs[-1].append(d[trainset,"train_accuracy",contexts])
                train_agent_values[-1].append(d[trainset,"train_value",contexts])

            accurs_mat = np.zeros([len(cont_vec),repeats])
            values_mat = np.zeros([len(cont_vec),repeats])
            train_accurs_mat = np.zeros([len(cont_vec),repeats])
            train_values_mat = np.zeros([len(cont_vec),repeats])
            
        for cont_i in range(len(cont_vec)):
            for trainset in range(repeats):
                accurs_mat[cont_i,trainset] = test_agent_accurs[trainset][cont_i][0]
                values_mat[cont_i,trainset] = test_agent_values[trainset][cont_i][0]/test_expert_value
                train_accurs_mat[cont_i,trainset] = train_agent_accurs[trainset][cont_i]
                train_values_mat[cont_i,trainset] = train_agent_values[trainset][cont_i]/train_expert_values[trainset*len(cont_vec) + cont_i]
                
        accurs_mean.append(accurs_mat.mean(axis=1))
        values_mean.append((100*values_mat).mean(axis=1))
        train_accurs_mean.append(train_accurs_mat.mean(axis=1))
        train_values_mean.append((100*train_values_mat).mean(axis=1))

        accurs_std.append(accurs_mat.std(axis=1))
        values_std.append((100*values_mat).std(axis=1))
        train_accurs_std.append(train_accurs_mat.std(axis=1))
        train_values_std.append((100*train_values_mat).std(axis=1))
    
    
    accurs_meanlin = []
    values_meanlin = []
    train_accurs_meanlin = []
    train_values_meanlin = []
    
    accurs_stdlin = []
    values_stdlin = []
    train_accurs_stdlin = []
    train_values_stdlin = []
                   
    for d in dlistBlin:
        train_agent_valueslin = []
        train_agent_accurslin = []
        test_agent_valueslin = []
        test_agent_accurslin = []
        
        for trainset in range(repeats):
            test_agent_valueslin.append([])
            test_agent_accurslin.append([])
            train_agent_valueslin.append([])
            train_agent_accurslin.append([])
            cont_veclin = d["context_nums"]
            test_expert_valuelin = d["test_value_expert"]
            train_expert_valueslin = d["train_value_expert"]
            TBSTSlin = True
            
            for contexts in cont_veclin:
                test_agent_accurslin[-1].append(d[trainset,"accuracy",contexts])
                test_agent_valueslin[-1].append(d[trainset,"test_value",contexts])
                train_agent_accurslin[-1].append(d[trainset,"train_accuracy",contexts])
                train_agent_valueslin[-1].append(d[trainset,"train_value",contexts])

            accurs_matlin = np.zeros([len(cont_veclin),repeats])
            values_matlin = np.zeros([len(cont_veclin),repeats])
            train_accurs_matlin = np.zeros([len(cont_veclin),repeats])
            train_values_matlin = np.zeros([len(cont_veclin),repeats])
            
        for cont_i in range(len(cont_veclin)):
            for trainset in range(repeats):
                accurs_matlin[cont_i,trainset] = test_agent_accurslin[trainset][cont_i][0]
                values_matlin[cont_i,trainset] = test_agent_valueslin[trainset][cont_i][0]/test_expert_valuelin
                train_accurs_matlin[cont_i,trainset] = train_agent_accurslin[trainset][cont_i]
                train_values_matlin[cont_i,trainset] = train_agent_valueslin[trainset][cont_i]/train_expert_valueslin[trainset*len(cont_veclin) + cont_i]
                
        accurs_meanlin.append(accurs_matlin.mean(axis=1))
        values_meanlin.append((100*values_matlin).mean(axis=1))
        train_accurs_meanlin.append(train_accurs_matlin.mean(axis=1))
        train_values_meanlin.append((100*train_values_matlin).mean(axis=1))

        accurs_stdlin.append(accurs_matlin.std(axis=1))
        values_stdlin.append((100*values_matlin).std(axis=1))
        train_accurs_stdlin.append(train_accurs_matlin.std(axis=1))
        train_values_stdlin.append((100*train_values_matlin).std(axis=1))  
    
    TCSTS = True             
    accurs_meanC = []
    values_meanC = []
    train_accurs_meanC = []
    train_values_meanC = []

    accurs_stdC = []
    values_stdC = []
    train_accurs_stdC = []
    train_values_stdC = []

    for d in dlistC:
        train_agent_valuesC = []
        train_agent_accursC = []
        test_agent_valuesC = []
        test_agent_accursC = []
        for trainset in range(repeats):
            test_agent_valuesC.append([])
            test_agent_accursC.append([])
            train_agent_valuesC.append([])
            train_agent_accursC.append([])
            cont_vecC = d["context_nums"]
            test_expert_valueC = d["test_value_expert"]
            train_expert_valuesC = d["train_value_expert"]
            TCSTS = True
            for contexts in cont_vecC:
                test_agent_accursC[-1].append(d[trainset,"accuracy",contexts])
                test_agent_valuesC[-1].append(d[trainset,"test_value",contexts])
                train_agent_accursC[-1].append(d[trainset,"train_accuracy",contexts])
                train_agent_valuesC[-1].append(d[trainset,"train_value",contexts])

            accurs_matC = np.zeros([len(cont_vecC),repeats])
            values_matC = np.zeros([len(cont_vecC),repeats])
            train_accurs_matC = np.zeros([len(cont_vecC),repeats])
            train_values_matC = np.zeros([len(cont_vecC),repeats])
            
        for cont_i in range(len(cont_vecC)):
            for trainset in range(repeats):
                accurs_matC[cont_i,trainset] = test_agent_accursC[trainset][cont_i][0]
                values_matC[cont_i,trainset] = test_agent_valuesC[trainset][cont_i][0]/test_expert_valueC
                train_accurs_matC[cont_i,trainset] = train_agent_accursC[trainset][cont_i]
                train_values_matC[cont_i,trainset] = train_agent_valuesC[trainset][cont_i]/train_expert_valuesC[trainset*len(cont_vecC) + cont_i]                
            
        accurs_meanC.append(accurs_matC.mean(axis=1))
        values_meanC.append((100*values_matC).mean(axis=1))
        train_accurs_meanC.append(train_accurs_matC.mean(axis=1))
        train_values_meanC.append((100*train_values_matC).mean(axis=1))
        
        accurs_stdC.append(accurs_matC.std(axis=1))
        values_stdC.append((100*values_matC).std(axis=1))
        train_accurs_stdC.append(train_accurs_matC.std(axis=1))
        train_values_stdC.append((100*train_values_matC).std(axis=1))
    
    values_mean_dim = []
    train_values_mean_dim = []
    accurs_mean_dim = []
    train_accurs_mean_dim =[]
    values_std_dim = []
    train_values_std_dim = []
    accurs_std_dim = []
    train_accurs_std_dim =[]
    
    for dd in range(len(dim_prob)):
        
        values_mean_dim_m = values_mean[dd][-1]
        train_values_mean_dim_m = train_values_mean[dd][-1]
        accurs_mean_dim_m = accurs_mean[dd][-1]
        train_accurs_mean_dim_m = train_accurs_mean[dd][-1]
        values_std_dim_m = values_std[dd][-1]
        train_values_std_dim_m = train_values_std[dd][-1]
        accurs_std_dim_m = accurs_std[dd][-1]
        train_accurs_std_dim_m = train_accurs_std[dd][-1]
            
        values_mean_dim.append(values_mean_dim_m)
        train_values_mean_dim.append(train_values_mean_dim_m)
        accurs_mean_dim.append(accurs_mean_dim_m)
        train_accurs_mean_dim.append(train_accurs_mean_dim_m)
        values_std_dim.append(values_std_dim_m)
        train_values_std_dim.append(train_values_std_dim_m)
        accurs_std_dim.append(accurs_std_dim_m)
        train_accurs_std_dim.append(train_accurs_std_dim_m)
        
    values_mean_dimlin = []
    train_values_mean_dimlin = []
    accurs_mean_dimlin = []
    train_accurs_mean_dimlin =[]
    values_std_dimlin = []
    train_values_std_dimlin = []
    accurs_std_dimlin = []
    train_accurs_std_dimlin =[]

    for dd in range(len(dim_prob)):
        values_mean_dim_mlin = values_meanlin[dd][-1]
        train_values_mean_dim_mlin = train_values_meanlin[dd][-1]
        accurs_mean_dim_mlin = accurs_meanlin[dd][-1]
        train_accurs_mean_dim_mlin = train_accurs_meanlin[dd][-1]
        values_std_dim_mlin = values_stdlin[dd][-1]
        train_values_std_dim_mlin = train_values_stdlin[dd][-1]
        accurs_std_dim_mlin = accurs_stdlin[dd][-1]
        train_accurs_std_dim_mlin = train_accurs_stdlin[dd][-1]
            
        values_mean_dimlin.append(values_mean_dim_mlin)
        train_values_mean_dimlin.append(train_values_mean_dim_mlin)
        accurs_mean_dimlin.append(accurs_mean_dim_mlin)
        train_accurs_mean_dimlin.append(train_accurs_mean_dim_mlin)
        values_std_dimlin.append(values_std_dim_mlin)
        train_values_std_dimlin.append(train_values_std_dim_mlin)
        accurs_std_dimlin.append(accurs_std_dim_mlin)
        train_accurs_std_dimlin.append(train_accurs_std_dim_mlin)        
    
    values_mean_dimC = []
    train_values_mean_dimC = []
    accurs_mean_dimC = []
    train_accurs_mean_dimC =[]
    values_std_dimC = []
    train_values_std_dimC = []
    accurs_std_dimC = []
    train_accurs_std_dimC =[]

    for dd in range(len(dim_prob)):
        values_mean_dim_mC = values_meanC[dd][-1]
        train_values_mean_dim_mC = train_values_meanC[dd][-1]
        accurs_mean_dim_mC = accurs_meanC[dd][-1]
        train_accurs_mean_dim_mC = train_accurs_meanC[dd][-1]
        values_std_dim_mC = values_stdC[dd][-1]
        train_values_std_dim_mC = train_values_stdC[dd][-1]
        accurs_std_dim_mC = accurs_stdC[dd][-1]
        train_accurs_std_dim_mC = train_accurs_stdC[dd][-1]
            
        values_mean_dimC.append(values_mean_dim_mC)
        train_values_mean_dimC.append(train_values_mean_dim_mC)
        accurs_mean_dimC.append(accurs_mean_dim_mC)
        train_accurs_mean_dimC.append(train_accurs_mean_dim_mC)
        values_std_dimC.append(values_std_dim_mC)
        train_values_std_dimC.append(train_values_std_dim_mC)
        accurs_std_dimC.append(accurs_std_dim_mC)
        train_accurs_std_dimC.append(train_accurs_std_dim_mC)
    
    values_mean_dim = np.asarray(values_mean_dim)
    train_values_mean_dim = np.asarray(train_values_mean_dim)
    accurs_mean_dim = np.asarray(accurs_mean_dim)
    train_accurs_mean_dim = np.asarray(train_accurs_mean_dim)
    values_std_dim = np.asarray(values_std_dim)
    train_values_std_dim = np.asarray(train_values_std_dim)
    accurs_std_dim = np.asarray(accurs_std_dim)
    train_accurs_std_dim = np.asarray(train_accurs_std_dim)
    
    values_mean_dimlin = np.asarray(values_mean_dimlin)
    train_values_mean_dimlin = np.asarray(train_values_mean_dimlin)
    accurs_mean_dimlin = np.asarray(accurs_mean_dimlin)
    train_accurs_mean_dimlin = np.asarray(train_accurs_mean_dimlin)
    values_std_dimlin = np.asarray(values_std_dimlin)
    train_values_std_dimlin = np.asarray(train_values_std_dimlin)
    accurs_std_dimlin = np.asarray(accurs_std_dimlin)
    train_accurs_std_dimlin = np.asarray(train_accurs_std_dimlin)
    
    values_mean_dimC = np.asarray(values_mean_dimC)
    train_values_mean_dimC = np.asarray(train_values_mean_dimC)
    accurs_mean_dimC = np.asarray(accurs_mean_dimC)
    train_accurs_mean_dimC = np.asarray(train_accurs_mean_dimC)
    values_std_dimC = np.asarray(values_std_dimC)
    train_values_std_dimC = np.asarray(train_values_std_dimC)
    accurs_std_dimC = np.asarray(accurs_std_dimC)
    train_accurs_std_dimC = np.asarray(train_accurs_std_dimC)
    
    fig1 = plt.figure(figsize=(6,4.5))
    p1 = fig1.add_subplot(111)
    p1.plot(dim_prob,values_mean_dim, color='r')
    p1.plot(dim_prob,values_mean_dimlin, color='g')
    p1.plot(dim_prob,values_mean_dimC, color='b')
    p1.errorbar(dim_prob,train_values_mean_dim,yerr=train_values_std_dim, fmt='none', color='r',capsize=5)
    p1.errorbar(dim_prob,train_values_mean_dimlin,yerr=train_values_std_dimlin, fmt='none', color='g',capsize=5)
    p1.errorbar(dim_prob,train_values_mean_dimC,yerr=train_values_std_dimC, fmt='none', color='b',capsize=5)
    p1.scatter(dim_prob,train_values_mean_dim, marker="o", facecolors='none', edgecolors='r', label="BC - deep train / test")
    p1.scatter(dim_prob,train_values_mean_dimlin, marker="o", facecolors='none', edgecolors='g', label="BC - linear train / test")
    p1.scatter(dim_prob,train_values_mean_dimC, marker="o", facecolors='none', edgecolors='b', label="COIRL (PSGD) train / test")
    p1.fill_between(dim_prob,values_mean_dim - values_std_dim, values_mean_dim + values_std_dim, color='r', alpha=0.1)
    p1.fill_between(dim_prob,values_mean_dimlin - values_std_dimlin, values_mean_dimlin + values_std_dimlin, color='g', alpha=0.1)
    p1.fill_between(dim_prob,values_mean_dimC - values_std_dimC, values_mean_dimC + values_std_dimC, color='b', alpha=0.1)
    p1.set_title("Value X Dimensions")
    p1.set_xlabel("Grid dimensions")
    p1.set_ylabel("% of Expert value")
    p1.set_ylim([50,102])
    p1.legend(loc=3)
    
    
    fig2 = plt.figure(figsize=(6,4.5))
    p2 = fig2.add_subplot(111)
    p2.plot(dim_prob,accurs_mean_dim, color='r')
    p2.plot(dim_prob,accurs_mean_dimlin, color='g')
    p2.plot(dim_prob,accurs_mean_dimC, color='b')
    p2.errorbar(dim_prob,train_accurs_mean_dim,yerr=train_accurs_std_dim, fmt='none', color='r',capsize=5)
    p2.errorbar(dim_prob,train_accurs_mean_dimlin,yerr=train_accurs_std_dimlin, fmt='none', color='g',capsize=5)
    p2.errorbar(dim_prob,train_accurs_mean_dimC,yerr=train_accurs_std_dimC, fmt='none', color='b',capsize=5)
    p2.scatter(dim_prob,train_accurs_mean_dim, marker="o", facecolors='none', edgecolors='r', label="BC - deep train / test")
    p2.scatter(dim_prob,train_accurs_mean_dimlin, marker="o", facecolors='none', edgecolors='g', label="BC - linear train / test")
    p2.scatter(dim_prob,train_accurs_mean_dimC, marker="o", facecolors='none', edgecolors='b', label="COIRL (PSGD) train / test")
    p2.fill_between(dim_prob,accurs_mean_dim - accurs_std_dim, accurs_mean_dim + accurs_std_dim, color='r', alpha=0.1)
    p2.fill_between(dim_prob,accurs_mean_dimlin - accurs_std_dimlin, accurs_mean_dimlin + accurs_std_dimlin, color='g', alpha=0.1)
    p2.fill_between(dim_prob,accurs_mean_dimC - accurs_std_dimC, accurs_mean_dimC + accurs_std_dimC, color='b', alpha=0.1)
    p2.set_title("Accuracy X Dimensions")
    p2.set_xlabel("Grid dimensions")
    p2.set_ylabel("% Hit")
    p2.set_ylim([10,102])
    p2.legend(loc=3)
    
    plt.show()
```
