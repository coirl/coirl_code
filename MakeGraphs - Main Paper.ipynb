```python
def make_online_drv():
    repeats = 5
    iters = 496
    dlistE = []
    for i in range(repeats):
        try:
            dlistE.append(load_obj("es_values"+str(i),'Online setting/Autonomous driving simulation/'))
            if iters == 0:
                iters = len(dlistE[0][0,"expert_values"])
        except:
            pass
    dlistB = []
    for i in range(repeats):
        try:
            dlistB.append(load_obj("psgd_values"+str(i),'Online setting/Autonomous driving simulation/'))
        except:
            pass
    dlistN = []
    for i in range(repeats):
        try:
            dlistN.append(load_obj("ew_values"+str(i),'Online setting/Autonomous driving simulation/'))
        except:
            pass

    expert_valuesE =    np.zeros((repeats,iters))
    agent_valuesE =     np.zeros((repeats,iters))
    contexts_seen_minE = np.inf

    contexts_seen_minE = iters - 1
    contexts_seen_minE = int(contexts_seen_minE)
    test_expert_valueE= np.zeros((repeats,contexts_seen_minE+5))
    test_agent_valueE = np.zeros((repeats,contexts_seen_minE+5))
    test_agent_lossE = np.zeros((repeats,contexts_seen_minE+5))
    
    TESTS = False
    for trainset in range(repeats):
        for d in dlistE:
            if (trainset,"test_value",90) in d:
                TESTS = True
                for contexts in range(0,contexts_seen_minE+1,5):
                    test_expert_valueE[trainset][contexts] = d["test_value"]
                    test_agent_lossE[trainset][contexts] = d[trainset,"test_loss",contexts]
                    test_agent_valueE[trainset][contexts] = d[trainset,"test_value",contexts]

    test_expert_valueE = test_expert_valueE[~np.all(test_expert_valueE == 0, axis=1)]
    test_agent_valueE = test_agent_valueE[~np.all(test_agent_valueE == 0, axis=1)]
    test_agent_lossE = test_agent_lossE[~np.all(test_agent_lossE == 0, axis=1)]

    contexts_seen_minB = np.inf
    contexts_seen_minB = iters - 1
    contexts_seen_minB = int(contexts_seen_minB)
    test_expert_valueB= np.zeros((repeats,contexts_seen_minB+2))
    test_agent_valueB = np.zeros((repeats,contexts_seen_minB+2))
    test_agent_lossB = np.zeros((repeats,contexts_seen_minB+2))

    TBSTS = False
    for trainset in range(repeats):
        for d in dlistB:
            if (trainset,"test_value",90) in d:
                TBSTS = True
                for contexts in range(0,contexts_seen_minB+1,5):
                    test_expert_valueB[trainset][contexts] = d["test_value"]
                    test_agent_lossB[trainset][contexts] = d[trainset,"test_loss",contexts]
                    test_agent_valueB[trainset][contexts] = d[trainset,"test_value",contexts]
                        
    test_expert_valueB = test_expert_valueB[~np.all(test_expert_valueB == 0, axis=1)]
    test_agent_valueB = test_agent_valueB[~np.all(test_agent_valueB == 0, axis=1)]
    test_agent_lossB = test_agent_lossB[~np.all(test_agent_lossB == 0, axis=1)]

    expert_valuesN = np.zeros((repeats,iters))
    agent_valuesN = np.zeros((repeats,iters))
    contexts_seen_minN = np.inf
    contexts_seen_minN = iters-1
    contexts_seen_minN = int(contexts_seen_minN)
    test_expert_valueN= np.zeros((repeats,contexts_seen_minN+5))
    test_agent_valueN = np.zeros((repeats,contexts_seen_minN+5))
    test_agent_lossN = np.zeros((repeats,contexts_seen_minN+5))

    TNSTS = False
    for trainset in range(repeats):
        for d in dlistN:
                if (trainset,"test_value",90) in d:
                    TNSTS = True
                    for contexts in range(0,contexts_seen_minN+5,5):
                        test_expert_valueN[trainset][contexts] = d["test_value"]
                        test_agent_lossN[trainset][contexts] = d[trainset,"test_loss",contexts]
                        test_agent_valueN[trainset][contexts] = d[trainset,"test_value",contexts]

    test_expert_valueN = test_expert_valueN[~np.all(test_expert_valueN == 0, axis=1)]
    test_agent_valueN = test_agent_valueN[~np.all(test_agent_valueN == 0, axis=1)]
    test_agent_lossN = test_agent_lossN[~np.all(test_agent_lossN == 0, axis=1)]

    if TESTS:
        fig1 = plt.figure(figsize=(6,4.5))
        p1 = fig1.add_subplot(111)
        
        test_expert_value_meanE = test_expert_valueE.mean(axis=0)
        test_expert_value_stdE = test_expert_valueE.std(axis=0)
        test_agent_value_meanE = (test_agent_valueE.mean(axis=0) / test_expert_value_meanE)
        test_agent_value_stdE = test_agent_valueE.std(axis=0) / test_expert_value_meanE
        test_agent_loss_meanE = test_agent_lossE.mean(axis=0)
        test_agent_loss_stdE = test_agent_lossE.std(axis=0)
        test_expert_value_meanB = test_expert_valueB.mean(axis=0)
        test_expert_value_stdB = test_expert_valueB.std(axis=0)
        test_agent_value_meanB = (test_agent_valueB.mean(axis=0) / test_expert_value_meanB)
        test_agent_value_stdB = test_agent_valueB.std(axis=0) /test_expert_value_meanB
        test_agent_loss_meanB = test_agent_lossB.mean(axis=0)
        test_agent_loss_stdB = test_agent_lossB.std(axis=0)
        test_expert_value_meanN = test_expert_valueN.mean(axis=0)
        test_expert_value_stdN = test_expert_valueN.std(axis=0)
        test_agent_value_meanN = (test_agent_valueN.mean(axis=0) / test_expert_value_meanN)
        test_agent_value_stdN = test_agent_valueN.std(axis=0) / test_expert_value_meanN
        test_agent_loss_meanN = test_agent_lossN.mean(axis=0)
        test_agent_loss_stdN = test_agent_lossN.std(axis=0)
        indices = list(range(0,contexts_seen_minE-5,5))

        p1.semilogy(indices,test_agent_loss_meanE[indices], label="ES", color='b')
        p1.semilogy(indices,test_agent_loss_meanB[indices], label="PSGD", color='r')
        p1.semilogy(indices,test_agent_loss_meanN[indices], label="EW", color='g')

        p1.fill_between(indices,test_agent_loss_meanE[indices] - test_agent_loss_stdE[indices], test_agent_loss_meanE[indices] + test_agent_loss_stdE[indices], color='b', alpha=0.1)
        p1.fill_between(indices,test_agent_loss_meanB[indices] - test_agent_loss_stdB[indices], test_agent_loss_meanB[indices] + test_agent_loss_stdB[indices], color='r', alpha=0.1)
        p1.fill_between(indices,test_agent_loss_meanN[indices] - test_agent_loss_stdN[indices], test_agent_loss_meanN[indices] + test_agent_loss_stdN[indices], color='g', alpha=0.1)
        p1.set_title("Loss X Contexts")
        p1.set_xlabel("Contexts")
        p1.set_ylabel("Loss")
        p1.tick_params('y', colors='k')
        plt.ylim([0.00001,0.02])
        p1.legend(loc=1)
        plt.show()

        fig2 = plt.figure(figsize=(6,4.5))
        p2 = fig2.add_subplot(111)
        p2.plot(indices, np.ones(len(indices)), label="Expert", color='k')
        p2.plot(indices,test_agent_value_meanE[indices], label="ES", color='b')
        p2.plot(indices,test_agent_value_meanB[indices], label="PSGD", color='r')
        p2.plot(indices,test_agent_value_meanN[indices], label="EW", color='g')

        p2.fill_between(indices,test_agent_value_meanE[indices] - test_agent_value_stdE[indices], test_agent_value_meanE[indices] + test_agent_value_stdE[indices], color='b', alpha=0.1)
        p2.fill_between(indices,test_agent_value_meanB[indices] - test_agent_value_stdB[indices], test_agent_value_meanB[indices] + test_agent_value_stdB[indices], color='r', alpha=0.1)
        p2.fill_between(indices,test_agent_value_meanN[indices] - test_agent_value_stdN[indices], test_agent_value_meanN[indices] + test_agent_value_stdN[indices], color='g', alpha=0.1)

        p2.set_title("Value X Contexts")
        p2.set_xlabel("Contexts")
        p2.set_ylabel("Value")
        p2.tick_params('y', colors='k')
        p2.legend(loc=4)
        plt.show()
    
    accurES = []
    accurPSG = []
    accurEW = []
    
    # plot accuracy:
    for i in range(repeats):
        try:
            accurES.append(np.load("./Online setting/Autonomous driving simulation/obj/accuracy_es_" + str(i) + ".npy"))
            
        except:
            pass
        
    for i in range(repeats):
        try:
            accurPSG.append(np.load("./Online setting/Autonomous driving simulation/obj/accuracy_psgd_" + str(i) + ".npy"))
            
        except:
            pass
        
    for i in range(repeats):
        try:
            accurEW.append(np.load("./Online setting/Autonomous driving simulation/obj/accuracy_ew_" + str(i) + ".npy"))
            
        except:
            pass

    if True:
        minlen_accurES = accurES[0].size
        for i in range(len(accurES)):
            if accurES[i].size < minlen_accurES:
                minlen_accurES = accurES[i].size +1
        
        accuracyES = np.zeros([len(accurES),minlen_accurES])
        for i in range(len(accurES)):
            accuracyES[i,:] = accurES[i][:minlen_accurES]
    
        minlen_accurPSG = accurPSG[0].size 
        for i in range(len(accurPSG)):
            if accurPSG[i].size < minlen_accurPSG:
                minlen_accurPSG = accurPSG[i].size + 1
                
        accuracyPSG = np.zeros([len(accurPSG),minlen_accurPSG])
        for i in range(len(accurPSG)):
            accuracyPSG[i,:] = accurPSG[i][:minlen_accurPSG]
            
        minlen_accurEW = accurEW[0].size
        for i in range(len(accurEW)):
            if accurEW[i].size < minlen_accurEW:
                minlen_accurEW = accurEW[i].size +1
                
        accuracyEW = np.zeros([len(accurEW),minlen_accurEW])
        for i in range(len(accurEW)):
            accuracyEW[i,:] = accurEW[i][:minlen_accurEW]
            
        accuracy_meanES = accuracyES.mean(axis=0)
        accuracy_stdES = accuracyES.std(axis=0)
        accuracy_meanPSG = accuracyPSG.mean(axis=0)
        accuracy_stdPSG = accuracyPSG.std(axis=0)
        accuracy_meanEW = accuracyEW.mean(axis=0)
        accuracy_stdEW = accuracyEW.std(axis=0)
        
        fig3 = plt.figure(
           figsize=(6,4.5))
        p3 = fig3.add_subplot(111)
        p3.plot(list(range(0,(minlen_accurES)*5,5)), 100*accuracy_meanES, label="ES", color='b')
        p3.fill_between(list(range(0,(minlen_accurES)*5,5)), 100*accuracy_meanES - 100*accuracy_stdES, 100*accuracy_meanES + 100*accuracy_stdES, color='b', alpha=0.1)
        
        p3.plot(list(range(0,(minlen_accurPSG)*5,5)), 100*accuracy_meanPSG, label="PSGD", color='r')
        p3.fill_between(list(range(0,(minlen_accurPSG)*5,5)), 100*accuracy_meanPSG - 100*accuracy_stdPSG, 100*accuracy_meanPSG + 100*accuracy_stdPSG, color='r', alpha=0.1)

        p3.plot(list(range(0,(minlen_accurEW)*5,5)), 100*accuracy_meanEW, label="EW", color='g')
        p3.fill_between(list(range(0,(minlen_accurEW)*5,5)), 100*accuracy_meanEW - 100*accuracy_stdEW, 100*accuracy_meanEW + 100*accuracy_stdEW, color='g', alpha=0.1)
        
        p3.set_title("Accuracy X Contexts")
        p3.set_xlabel("Contexts")
        p3.set_ylabel("% Hit")
        p3.legend(loc=4)
        plt.show()

    
```

```python
def make_offline_med_deep():
    repeats = 5
    iters = 105
    dlistB = []
    for i in range(repeats):
        try:
            dlistB.append(load_obj("psgd_values"+str(i),'Offline setting/Dynamic treatment regime/'))
        except:
            pass
    expert_valuesB = np.zeros((repeats, iters))
    agent_valuesB = np.zeros((repeats, iters))

    for trainset in range(repeats):
        for d in dlistB:
            if (trainset, "expert_values") in d:
                expert_valuesB[trainset] = d[trainset, "expert_values"][:iters]
                agent_valuesB[trainset] = d[trainset, "agent_values"][:iters]

    contexts_seen_minB = iters - 1
    test_expert_valueB = np.zeros((repeats, contexts_seen_minB + 2))
    test_agent_valueB = np.zeros((repeats, contexts_seen_minB + 2))
    test_agent_lossB = np.zeros((repeats, contexts_seen_minB + 2))

    TBSTS = False
    for trainset in range(repeats):
        for d in dlistB:
            if (trainset, "test_value", 195) in d:
                TBSTS = True
                for contexts in range(0, contexts_seen_minB + 1, 5):
                    test_expert_valueB[trainset][contexts] = d["test_value"]
                    test_agent_lossB[trainset][contexts] = d[trainset, "test_loss", contexts]
                    test_agent_valueB[trainset][contexts] = d[trainset, "test_value", contexts]
                    
    expert_valuesB = expert_valuesB[~np.all(expert_valuesB == 0, axis=1)]
    agent_valuesB = agent_valuesB[~np.all(agent_valuesB == 0, axis=1)]
    test_expert_valueB = test_expert_valueB[~np.all(test_expert_valueB == 0, axis=1)]
    test_agent_valueB = test_agent_valueB[~np.all(test_agent_valueB == 0, axis=1)] / test_expert_valueB[0]
    test_agent_lossB = test_agent_lossB[~np.all(test_agent_lossB == 0, axis=1)]

    expert_values_stdB = expert_valuesB.std(axis=0)
    agent_values_stdB = agent_valuesB.std(axis=0)
    expert_values_meanB = expert_valuesB.mean(axis=0)
    agent_values_meanB = agent_valuesB.mean(axis=0)

    TESTS = True
    if TESTS:
        test_expert_value_meanB = (100*test_expert_valueB).mean(axis=0)
        test_expert_value_stdB = (100*test_expert_valueB).std(axis=0)
        test_agent_value_meanB = (100*test_agent_valueB).mean(axis=0)
        test_agent_value_stdB = (100*test_agent_valueB).std(axis=0)
        test_agent_loss_meanB = test_agent_lossB.mean(axis=0)
        test_agent_loss_stdB = test_agent_lossB.std(axis=0)
        indices = list(range(0, contexts_seen_minB, 5))

        fig1 = plt.figure(figsize=(6, 4.5))
        p1 = fig1.add_subplot(111)
        p1.plot(indices, test_agent_loss_meanB[indices], label="PSGD", color='r')
        p1.fill_between(indices,test_agent_loss_meanB[indices] - test_agent_loss_stdB[indices], test_agent_loss_meanB[indices] + test_agent_loss_stdB[indices], color='r', alpha=0.1)
        
        p1.set_title("Loss X Iterations")
        p1.set_xlabel("Iterations")
        p1.set_ylabel("Loss")
        p1.tick_params('y', colors='k')
        p1.legend(loc=1)

        plt.show()

        fig2 = plt.figure(figsize=(6, 4.5))
        p2 = fig2.add_subplot(111)
        p2.plot(indices, test_agent_value_meanB[indices], label="PSGD", color='r')
        p2.fill_between(indices,test_agent_value_meanB[indices] - test_agent_value_stdB[indices], test_agent_value_meanB[indices] + test_agent_value_stdB[indices], color='r', alpha=0.1)

        p2.set_title("Value X Iterations")
        p2.set_xlabel("Iterations")
        p2.set_ylabel("Value")
        p2.legend(loc=4)
        plt.show()

        accurPSG = []

        for i in range(repeats):
            try:
                accurPSG.append(np.load("./Offline setting/Dynamic treatment regime/obj/accuracy_psgd_" + str(i) + ".npy"))

            except:
                print("PSGD", i)
                pass

        minlen_accurPSG = accurPSG[0].size
        for i in range(len(accurPSG)):
            if accurPSG[i].size < minlen_accurPSG:
                minlen_accurPSG = accurPSG[i].size + 1

        accuracyPSG = np.zeros([len(accurPSG), minlen_accurPSG])
        for i in range(len(accurPSG)):
            accuracyPSG[i, :] = accurPSG[i][:minlen_accurPSG]

        accuracy_meanPSG = (100*accuracyPSG).mean(axis=0)
        accuracy_stdPSG = (100*accuracyPSG).std(axis=0)
        fig3 = plt.figure(figsize=(6, 4.5))
        p3 = fig3.add_subplot(111)

        p3.plot(list(range(0, (minlen_accurPSG) * 5, 5)), accuracy_meanPSG, label="PSGD", color='r')
        p3.fill_between(list(range(0,(minlen_accurPSG)*5,5)), accuracy_meanPSG - accuracy_stdPSG, accuracy_meanPSG + accuracy_stdPSG, color='r', alpha=0.1)

        p3.set_title("Accuracy X Iterations")
        p3.set_xlabel("Iterations")
        p3.set_ylabel("% Hit")
        p3.set_ylim(15, 80)
        p3.legend(loc=4)
        plt.show()

        figg, ax1 = plt.subplots(figsize=(7.2, 4.75))

        color = 'darkgreen'
        ax1.set_xlabel('Iterations')
        ax1.set_ylabel('% of Expert') 
        ax1.set_ylim([25, 100])
        ax1.plot(indices, accuracy_meanPSG[:len(indices)], color='g', label='Accuracy', linestyle='--')
        ax1.plot(indices, test_agent_value_meanB[indices], color=color, label='Value', linestyle='-')
        ax1.tick_params(axis='y', labelcolor=color)
        ax1.plot([], [], color='r', label='Loss')
        plt.legend(loc=4, bbox_to_anchor=(1, 0.1))
        ax2 = ax1.twinx() 

        color = 'tab:red'
        ax2.set_ylabel('Loss') 
        ax2.plot(indices, test_agent_loss_meanB[indices], color='r')
        ax2.tick_params(axis='y', labelcolor=color)
        ax2.set_ylim([0, 0.02])
        figg.tight_layout() 
        plt.title("Paper graph")
        plt.show()
        
        print("Final performance:")
        print("Loss mean: ", (test_agent_loss_meanB[indices])[-1])
        print("Loss std: ", (test_agent_loss_stdB[indices])[-1])
        print("Value mean: ", (test_agent_value_meanB[indices])[-1])
        print("Value std: ", (test_agent_value_stdB[indices])[-1])
        print("Accuracy mean: ", accuracy_meanPSG[len(indices)-1])
        print("Accuracy std: ", accuracy_stdPSG[len(indices)-1])
```
